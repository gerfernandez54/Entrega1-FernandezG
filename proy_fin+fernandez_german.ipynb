## **Introduccion**

---

La conexión a internet es una necesidad fundamental en la vida moderna y es por eso que la calidad del servicio es de suma importancia. En este proyecto, nos enfocamos en la segmentación de clientes/equipos para mejorar el índice de experiencia del usuario (CEI). La red de internet se distribuye por todo el país a través de dos sentidos de fibra óptica, uno principal y otro secundario, que llegan al Repartidor General (RP) de cada empresa proveedora de internet. Desde allí, se distribuye a través de nodos que abarcan una promedio de 3 o 4 manzanas en un barrio, dependiendo de la cantidad y tipo de clientes.

Cada nodo es una aglomeración de clientes que solicitan su servicio de internet. El Nodo Óptico (NO) transforma la luz en RF y distribuye la señal por la red de calle hasta el Tap, que es el equipo pasivo en el que se conecta al cliente y se realiza la bajada hasta el techo, donde se encuentra el cableado interno del domicilio. Cada cliente tiene un equipo módem que le permite conectarse a internet y recibir una dirección IP asignada por los CMTS.

El índice de experiencia del usuario (CEI) es una medida de la calidad del servicio que se proporciona a cada cliente. Se calcula en base a una fórmula que tiene en cuenta varios factores técnicos y se utiliza para evaluar el desempeño de cada equipo en el parque desplegado. El CEI se mide semanalmente y se utiliza para identificar los equipos que funcionan bien (Promotores), los que funcionan mal (Detractores) y los que están en un punto intermedio (Neutros).

El objetivo de este proyecto es segmentar a los clientes/equipos según su CEI para poder mejorar la calidad del servicio que se les proporciona. Para lograr esto, se utilizará un histórico de 12 semanas de CEI para predecir el valor aproximado que tendrá en la semana siguiente. De esta manera, se podrá trabajar de manera preventiva para mejorar el CEI de la semana próxima y así mantener a los clientes Promotores satisfechos. Los Detractores y Neutros también serán un foco de atención para mejorar la calidad del servicio que se les proporciona y mantenerlos satisfechos. En resumen, este proyecto se enfoca en mejorar la calidad del servicio de internet a través de la segmentación de clientes/equipos según su CEI y la implementación de medidas preventivas para mantener a los clientes satisfechos.

## **Analisis de variables tecnicas**

Una de las problematicas mas grandes que tenemos es saber a clientes apuntar para mejorar su CEI. Para ello debemos sectorizar las principales tres causas que generan el decrecimiento de CEI. Estas son la Salud de Red, la Estabilidad (se entiende por Estabilidad, el comportamiento propio de un modem de mantenerse estable en potencia electrica) y por ultimo el Daño (se entiende por Daño cuando hay una falla de una zona grande que afecta a muchos clientes).
Dicho esto, el primer paso es evaluar los clientes que tienen afectacion por los tres factores simultaneos. Son los principales a recuperar valor por CEI.  

```python
nodos_filtrados = df.loc[(df['observado_saludred'] == 100) & (df['observado_estabilidad'] == 100) & (df['observado_dano_open'] == 100)]
nodos_agrupados = nodos_filtrados.groupby('hub').size().sort_values(ascending=False)
nodos_agrupados
```

## **Clustering**

```python
df_Sdr_CEI.isnull().sum()
```

```python
df_Sdr_CEI = df_Sdr_CEI.fillna(0)
```

```python
# Seleccionar las columnas relevantes para el análisis de clustering
data = df_Sdr_CEI[['cei', 'observado_saludred']]

# Estandarizar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Aplicar PCA para reducción de dimensionalidad
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Encontrar el número óptimo de clusters utilizando Elbow Method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(pca_data)
    inertia.append(kmeans.inertia_)

# Visualizar Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Determinar el Número de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inertia')
plt.show()

# Elegir el número óptimo de clusters y aplicar KMeans
n_clusters = 3  # Seleccionar el número de clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
clusters = kmeans.fit_predict(pca_data)

# Agregar información de clusters al DataFrame original
df_Sdr_CEI['cluster'] = clusters

# Visualizar los clusters en el espacio reducido de PCA
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters, cmap='viridis')
plt.title('Clusters en el Espacio Reducido de PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
